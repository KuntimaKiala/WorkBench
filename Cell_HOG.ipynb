```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Step One
Charging the the image and tranforming it from RGB image to grayscale image

```python
image = cv2.imread('bat.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print image.shape
```

# Step Two 
sobel filter, to perform de first derivative and vsualizing

```python
gx =cv2.Sobel(gray,cv2.CV_32F,1,0, 1)
gy =cv2.Sobel(gray,cv2.CV_32F,0,1, 1)
mag,angle =cv2.cartToPolar(gx, gy, angleInDegrees =1 )

mag[mag-gray < -0.059*mag ]=0
#angle = np.int32(16*angle/(2*np.pi))
plt.imshow(mag)
plt.show()

```

# Step Three 
Creating the cells and making the histogram of each  cell

```python
cells_mag = [np.hsplit(row,3) for row in np.vsplit(mag,3)]
cells_ang = [np.hsplit(row,3) for row in np.vsplit(angle,3)]
histo = []
m =[]
a =[]



for cell_m in cells_mag :
    k = len(cell_m)
    for index in range(k) :
        m.append(cell_m[index])
        
for cell_a in cells_ang :
    k = len(cell_a)
    for index in range(k) :
        a.append(cell_a[index])

z_mag = np.array(m)
z_ang =np.array(a)
print z_mag.shape

for i in range(z_mag.shape[0]) :
    plt.hist(z_ang[i].ravel(), bins = [0,20,40,60,80,100,120,140,160,180], weights = z_mag[i].ravel())
    plt.show()
    

#hists = [np.bincount(b.ravel(),m.ravel(),16) for b, m in zip(z_mag , z_ang)]




```
